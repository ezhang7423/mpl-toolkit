```python
############################################
# science lineplot
import scienceplots
import matplotlib.pyplot as plt

plt.style.use("science")

line1_xy = ([0, 1, 2], [1, 2, 3])
line2_xy = ([0, 1, 2], [0, 1, 3])
plt.plot(*line1_xy)
plt.plot(*line2_xy)
plt.title("the  quick fox jumped over the lazy dog 2 $x = 2$")
plt.legend(["line 1", "line 2"])
plt.xlabel("x-axis $\mu$")
plt.ylabel("y-axis $\psi$")
plt.show()
```

```python
xy = ([0, 0, 0, 1, 2], [0.0, 1.2, 1, 2, 3])
plt.scatter(*xy)
plt.title("so this is good \n the quick fox jumped over the lazy dog 2 $x = 2$")
plt.legend(["scatter 1"])
plt.xlabel("x-axis $\mu$")
plt.ylabel("y-axis $\psi$")
plt.show()
```
